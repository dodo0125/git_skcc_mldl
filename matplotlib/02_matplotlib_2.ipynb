### 1. Bar plot
- plt.plot을 사용하는 대신 plt.bar를 사용합니다.

````python
y = np.random.randint(10, size=5)
print(y)
x = range(5)
plt.bar(x, y)
plt.show()
````

에러 데이터를 추가

````python
y = np.random.randint(5, 10, size=5)
x = range(5)
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.bar(x, y, yerr=error, align='center', alpha=0.6)
plt.show()
````

가로 bar 그래프

````python
y = np.random.randint(5, 10, size=5)
x = range(5)
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.barh(x, y, xerr=error, align='center', alpha=0.6)
plt.show()
````

두개의 bar 그래프를 겹침
- 나중에 그려진 그래프가 먼저 그려진 그래프 위로 올라감

````python
y1 = np.random.randint(5, 10, size=5)
y2 = np.random.randint(1, 5, size=5)
print(y1, y2)
plt.subplots()
x = range(5)
plt.bar(x, y1)
plt.bar(x, y2)

# plt.tight_layout()
plt.show()
````

### 2. pie plot

### 3. sactter plot

### 4. histogram

- bins: 데이터의 구간을 몇개로 나눌지 설정
- 50개의 구간을로 잘라 해당되는 데이터의 수를 카운트한 후에 y축의 값을 normalized

````python
x1 = np.random.rand(1000)
plt.subplot(211)
plt.hist(x1, bins=100, density=True)

x2 = np.random.randn(1000)
plt.subplot(212)
plt.hist(x2, bins=50, density=True)

plt.show()
````

````python
x = np.random.randint(0, 100, size=100)
print(x, np.bincount(x))
plt.subplot(311)
plt.hist(x)
plt.subplot(312)
plt.hist(x, bins=100)
plt.subplot(313)
plt.hist(x, density=True)

plt.show()
````

#### seaborn 적용 전의 그래프

````python
def sinplot():
    x = np.linspace(0, 10, 101)  # 0-10 101 등분
    for i in range(6):
        plt.plot(x, np.sin(i + x))  # 60도씩 sin이 이동된 데이터를 그려줌


sinplot()
````

#### 6.1 붓꽃 데이터
- lmplot, pairplot 실습
- sepal : 꽃받침
- petal : 꽃잎
- 꽃받침과 꽃잎의 넓이과 길이로 붓꽃 데이터의 종을 구분하기 위한 그래프를 시각화

````python
iris = sns.load_dataset("iris")
iris.tail()
````

'setosa', 'versicolor', 'virginica' 3가지 종이 있음을 확인

````python
iris["species"].unique()
````

꽃받침의 길이와 넓이로 종을 구분
- setosa 구분은 가능하지만 versicolor, virginica는 구분이 어려움

````python
sns.lmplot("sepal_length", "sepal_width", data=iris, hue='species', fit_reg=False)
plt.show()
````

꽃잎의 길이와 넓이로 종을 구분
- setosa 구분이 확실하게 가능하고 versicolor, virginica는 구분도 어느정도 가능
- 꽃받침으로 종을 구분하는것보다 꽃잎으로 종을 구분하는것이 더 쉬움

````python
sns.lmplot("petal_length", "petal_width", data=iris, hue='species', fit_reg=False)
plt.show()
````

pairplot 모든 컬럼을 lmplot으로 그림

````python
sns.pairplot(iris[["petal_length", "petal_width", "species"]], hue="species")
plt.show()
````

#### 6.2 팁 데이터
- jointplot, heatmap 실습
- sex : 성별
- time : 식사시간

````python
tips = sns.load_dataset("tips")
tips.tail(5)
````

##### 6.2.1 식사시간과 성별에 따른 팁의 총량을 시각화
- parameter
    - annot : 수치 정보 출력
    - fmt : 수치데이터 포멧 ".2f" 소수 둘째 자리
    - as_cmap: 스무스하게 표현
- [light_palette](http://seaborn.pydata.org/generated/seaborn.light_palette.html)

시간별 성별 팁의 총량

````python
tips_sum = tips.pivot_table(index="time", columns="sex", values="tip", aggfunc=np.sum)
    
tips_sum
````

결론 : 점심시간 보다 저녁시간에 팁이 많고, 저녁시간에 남자손님에게 받는 팁이 가장 많다.

````python
sns.heatmap(tips_sum, cmap=sns.light_palette("purple", as_cmap=True), annot=True, fmt=".2f")
    
plt.show()
````

#### 6.2.2 날짜와 흡연여부에 따른 팁을 주는 정도
- 날짜와 흡연여부에 따라서 가장 효율이 좋은 테이블을 선점하여 서빙을 하는게 좋을까?

날짜와 흡연여부에 따른 팁의 총액수

````python
tips_amount = tips.pivot_table(index="smoker", columns="day", values="tip", aggfunc="sum")
    
tips_amount
````

날짜와 흡연여부에 따른 고객의 수

````python
tips_count = tips.pivot_table(index="smoker", columns="day", values="tip", aggfunc="count")

tips_count
````

- 사람 수에 따른 평균 팁
    - 일요일 점심에 담배를 피우는 사람이 팁을 가장 많이 준다.

````python
tips_result = round(tips_amount / tips_count, 2)
tips_result
````

결론 : 일요일 점심에 담배를 피우는 사람의 테이블을 맡으면 적게 일하고 많은 팁을 받을수 있다.

````python
sns.heatmap(tips_result, cmap=sns.light_palette(
    "seagreen", as_cmap=True), annot=True, fmt=".2f")
plt.show()
````

##### 6.2.3 전체 계산 금액과 팁의 상관관계

````python
import scipy.stats as stats
s = sns.jointplot("total_bill", "tip", data=tips)
pearsonr, p_value = stats.pearsonr(tips["total_bill"], tips["tip"])
print("pearsonr : {}, p_value : {}".format(pearsonr, p_value))
plt.show()
````

#### 6.3 월별, 년도별 여객기 승객수 
- heatmap 실습

````python
# 여객기 승객수 : heatmap
flights = sns.load_dataset("flights")
flights.tail(2)
````

````python
flights_pivot = flights.pivot("month", "year", "passengers")
flights_pivot
````

여름시즌에 승객이 많고, 시간이 지날수록 점점 승객이 많아짐을 확인

````python
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.show()
````
