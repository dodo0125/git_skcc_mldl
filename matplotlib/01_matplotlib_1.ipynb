단축 옵션으로 한번에 설정
- 색상, 마커, 라인 스타일을 설정할수 있습니다.

```python
data = np.random.randint(10, size=10)
plt.plot(data, 'r*:')
plt.show()
```

커스텀하게 설정
- 단축 옵션으로 설정하는것보다 더 많은 설정을 할수 있습니다.

```python
data = np.random.randint(10, size=10)
plt.plot(data, c="b", ls="--", marker="o", lw=5,
         ms="10", mec="r", mew="3", mfc="g")
plt.show()
```

### 5. limit 설정
- x축과 y축의 범위를 정의 할수 있습니다.

xlim, ylim 설정

```python
# x 축 : -5 ~ 20 까지의 범위, y 축 : -10 ~ 30 까지의 범위
data = np.random.randint(10, size=10)
plt.xlim(-5, 20)
plt.ylim(-10, 30)
print(data)
plt.plot(data)
plt.show()
```

### 6. tick 설정
- 그래프에서의 x축과 y축의 값을 설정

```python
data = np.random.randint(5, size=5)
plt.plot(data)
plt.show()
```

```python
# x축과 y축의 눈금 데이터를 설정
print(data)
plt.plot(data)
plt.xticks(range(0, 10))
plt.yticks([1, 2])
plt.show()
```

### 8. 멀티 라인 그리기

하나의 plot 함수로 그리기

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
data3 = np.random.randint(5, size=5)
xticks = range(0, 5)
print(data1, data2, data3)

plt.xticks(range(0, 5))
plt.plot(xticks, data1, 'ro-', xticks, data2, 'g^:', xticks, data3, 'b*--')
plt.show()
```

여러개의 plot 함수로 그리기

```python
xticks = range(0, 5)
print(data1, data2, data3)

plt.plot(xticks, data1)
plt.plot(xticks, data2)
plt.plot(xticks, data3)
plt.show()
```

### 9. legend (범례)

- center left : 6

가장 적절한 위치에 자동으로 범례 배치

### 10. title 추가
- 그래프의 타이틀을 설정할수 있습니다.
- 기본 폰트가 한글폰트가 없는 `sans-serif`로 되어 있기 때문에 한글을 사용하면 글자가 깨집니다.

```python
data = np.random.randint(5, size=5)
plt.plot(data)
plt.xlabel("xtick")
plt.ylabel("ytick")
plt.title("Title")
plt.show()
```

폰트를 한글로 사용하면 폰트가 깨집니다.

```python
data = np.random.randint(5, size=5)
plt.plot(data)
plt.xlabel("x축")
plt.ylabel("y축")
plt.title("타이틀")
plt.show()
```

#### 한글 폰트 설정  방법 2
- 폰트 경로를 지정
- mac => /Library/Fonts/AppleMyungjo.ttf
- windows => C:/Windows/fonts/gulim.ttc

```python
import matplotlib.font_manager as fm
font_location = "C:/Windows/fonts/gulim.ttc" # windows
font_location = "/Library/Fonts/AppleMyungjo.ttf" # mac 카탈리나 이전 버전
font_location = "/System/Library/Fonts/Supplemental/AppleMyungjo.ttf" # mac 카탈리나 버전
font_name = fm.FontProperties(fname=font_location).get_name()
font_name
```

```python
# 폰트 적용
mpl.rc("font", family=font_name)
```

#### 12.1 그래프 크기 조절

#### 12.2 여러개 그래프 출력

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
print(data1, data2)

plt.subplot(2, 1, 1)  # 2*1에서 1번 위치
plt.plot(data1)

plt.subplot(2, 1, 2)  # 2*1에서 2번 위치
plt.plot(data2)

plt.show()
```

```python
data = [np.random.randint(5, size=5) for _ in range(4)]

print(data)

for idx in range(4):
    print("2*2에서 {}번 위치 : {}".format(idx + 1, 221 + idx))
    plt.subplot(221 + idx)
    plt.plot(data[idx])

plt.show()
```
