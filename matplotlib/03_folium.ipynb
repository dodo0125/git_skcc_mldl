#### 2. covid19
- `https://www.kaggle.com/davidbnn92/weather-data-for-covid19-data-analysis`
- `https://www.kaggle.com/davidbnn92/weather-data-for-covid19-data-analysis#training_data_with_weather_info_week_4.csv`

```python
df = pd.read_csv("training_data_with_weather_info_week_4.csv")
df.tail(2)
```

```python
df_1 = df.groupby(["Lat", "Long"]).size().reset_index(name="counts")
df_1.tail(2)
```

```python
map_obj = folium.Map(location=[0, 0], zoom_start=2)

for idx, data in df_1.iterrows():
    
    folium.CircleMarker([data.Lat, data.Long], radius=data.counts/8, color='red', fill=True).add_to(map_obj)
    
map_obj
```
